```python
import requests
import json
# para coletar data e hora
from datetime import datetime

#vamos colocar nosso endpoint
url = 'https://script.google.com/macros/s/AKfycbzZ8Drf5l2WcCDunuwlPzPKb3NMx_3KqoTxMcbzgddS_Iv7Rs8C7T5gxIxKSYVtSf-vQw/exec'

#agora vamos criar a função que envia os dados para o excel
def enviar_dados(placa, carro, funcionario):
  # vou colocar aqui a data e hora
  data_e_hora = datetime.now().strftime('%d/%m/%y %H:%M:%S')
  #%d - dia
  #%m - mês
  #%y - ano
  #%H - hora
  #%M - minutos
  #%S - segundo
  # datetime.now() - salva o momento que o código é executado
  modelo = ''
  parametros = {
      'data_hora': data_e_hora,
      'placa': placa,
      'carro':modelo,
      'funcionario':funcionario
  }
  print('Enviando dados para a tabela')

  resposta = requests.post(url, json=parametros)

  if resposta.status_code == 200:
    print('Dados prontos para envio')
  else:
    print(f'Erro ao enviar os dados: {resposta.status_code}')
```

```python
while True:
    print('-'*50)
    placa = input('Informe a placa do carro: ')
    modelo = input('Informe o modelo do carro: ')
    funcionario = input('Informe o funcionario que registrou: ')

    enviar_dados(placa, modelo, funcionario)

    op = input('Deseja sair? S/N')
    if op == 's':
      break
```
