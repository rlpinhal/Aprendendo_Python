```python
#comandos basicos de um banco de dados

%%sql
create table minhaplaylist(
  id_minhaplaylist INTEGER PRIMARY KEY AUTOINCREMENT,
  musica TEXT NOT NULL,
  artista TEXT NOT NULL
);
```

```python
print('----- MINHA PLAYLIST -----')
while True:
  musica = input('Informe o nome da música: ')
  artista = input('Informe o artista: ')
# Executa apenas esta linha como SQL, puxando as variáveis do Python
# vamos chamar nosso paas para salvar as informações
# insert into nome_da_tabela (campos) values (valores)
  %sql INSERT INTO minhaplaylist (musica, artista) VALUES (:musica, :artista)
  print('Informações adicionadas para o banco de dados')

  op = input('Quer adicionar nova musica? S/N: ')
  if op.lower() == 'n':
    break
```

```python
%%sql
SELECT * FROM minhaplaylist
```
