```python
import requests

# vamos usar a biblioteca requests para ver o código do status da api
# status_code
# 200 - esta funcionando --- 404 - arquivo não encontrado

#toda api tem uma propriedade chamada de end-point, que é a URL de onde iremos extrair as informações

url = 'https://api.open-meteo.com/v1/forecast'

# toda api precisa de parametros para funcionar
# precisamos definir os parametros para seu uso
latitude = float(input('Informe a Latitude desejada: '))
longitude = float(input('Informe a Longitude desejada: '))

parametros = {
    "latitude": latitude,
    "longitude": longitude,
    "current_weather": True
}

#vamos fazer uma requisição para coletar os dados do sistema
# nao posso só pegar (get) a informação
# temos que requisitar ela
resposta = requests.get(url, params=parametros)
# para eu coletar algo, chamamos a biblioteca requests com o atributo get(onde, o que)
# request.get(URL,params = PARAMETROS)

if resposta.status_code == 200:

  #vamos fzer funcinar
  #API é melhor com JSON
  #vamos colocar a resposta em um JSON
  dados = resposta.json()

  #vamos falar o que queremos da API
  temperatura = dados['current_weather']['temperature']
  velocidade_vento = dados['current_weather']['windspeed']

  print(f'A temperatura é: {temperatura}ºC')
  print(f'A velocidade do vento é {velocidade_vento} km/h')

# se a api caiu ou não funciona
else:
  print(f'Erro ao acessar a API - {resposta.status_code}')
```
