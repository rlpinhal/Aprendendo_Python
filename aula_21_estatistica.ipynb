```python
import statistics
semana_1 = [15,22,18,25,17,21]
semana_2 = [20,25,19,28,23,30]

print(f"A média da semana 1 é {statistics.mean(semana_1)}, e a média da semana 2 é {statistics.mean(semana_2)}")
print(f"A mediana da semana 1 é {statistics.median(semana_1)}, e a mediana da semana 2 é {statistics.median(semana_2)}")
print(f"A moda da semana 1 é {statistics.mode(semana_1)}, e a moda da semana 2 é {statistics.mode(semana_2)}")
print(f"O desvio padrão da semana 1 é {round(statistics.pstdev(semana_1),2)}, e a média da semana 2 é {round(statistics.pstdev(semana_2),2)}")

```
