```python
#definir as variaveis

jogador1 = input('Informe o nome do jogador 1: ')
jogador2 = input('Informe o nome do jogador 2: ')

pontuacao1= 0
pontuacao2= 0

movimento1 = None
movimento2 = None
```

```python
#definir as ações - funções

#funcao de escolher jogada
def escolher_jogada():
  print('='*20)
  print('As jogadas que temos são:')
  print('Pedra')
  print('Papel')
  print('Tesoura')
  print('='*20)
  print('DIGITE O TEXTO DA OPÇÃO: ')

  movimento1 = input(f'{jogador1}, informe sua jogada: ').lower()
  movimento2 = input(f'{jogador2}, informe sua jogada: ').lower()
  return movimento1,movimento2

#funcao de jogar. Adicionamos as variáveis de fora dentro da função para que ela consiga alterar a medida que o jogo for ocorrendo.
def jogar(movimento1,movimento2,pontuacao1,pontuacao2):
  print(f'{jogador1} escolheu {movimento1}')
  print(f'{jogador2} escolheu {movimento2}')

# se o jogador 1 ganhar
  if((movimento1 == 'pedra' and movimento2 == 'tesoura') or (movimento1 == 'papel' and movimento2 == 'pedra') or (movimento1 == 'tesoura' and movimento2 == 'papel')):
    print(f'{jogador1} escolheu {movimento1} e ganhou do {jogador2} que escolheu {movimento2}')
    pontuacao1 += 1
    print(f'A pontuação de {jogador1} é {pontuacao1}')

# se o jogador 2 ganhar
  elif((movimento2 == 'pedra' and movimento1 == 'tesoura') or (movimento2 == 'papel' and movimento1 == 'pedra') or (movimento2 == 'tesoura' and movimento1 == 'papel')):
    print(f'{jogador2} escolheu {movimento2} e ganhou do {jogador1} que escolheu {movimento1}')
    pontuacao2 += 1
    print(f'A pontuação de {jogador2} é {pontuacao2}')

  else:
    print('Empatou')

  return pontuacao1, pontuacao2

def regras():
  print('Pedra ganha de tesoura')
  print('Papel ganha de pedra')
  print('Tesoura ganha de papel')

```

```python
print('Vamos iniciar o jogo?')
print('1 - Regras')
print('2 - Jogar')
escolha = input('Qual opção desejada? 1 ou 2? ')

if escolha == '1':
  regras()
elif escolha == '2':
  movimento1, movimento2 = escolher_jogada()
  pontuacao1,pontuacao2 = jogar(movimento1, movimento2, pontuacao1, pontuacao2)
else:
  print('Opção não encontrada')
```
