```python
import numpy as np

#numpy é uma biblioteca de funções matemáticas
```

```python
# sem usar o numpy
# matemárica básica:

numero1 = float(input('informe o primeiro número: '))
numero2 = float(input('informe o segundo número: '))

# soma
soma = numero1 + numero2
print(soma)

#subtração
subtracao = numero1 - numero2
print(subtracao)

# divisão
divi = numero1/numero2
print(divi)

#multiplicação
multi = numero1 * numero2
print(multi)
```

```python
# matemática básica (mesmos cálculos do bloco anterior)
numero1 = float(input('informe o primeiro número: '))
numero2 = float(input('informe o segundo número: '))

#soma pode ser sum ou add
print(f'soma: {np.add(numero1, numero2)}')
#ou "sum" - só para números inteiros:
# para usar sum, teria que mudar então primeiro:
#numero1 = int(input('informe o primeiro número: '))
#numero2 = int(input('informe o segundo número: '))
# a fórmula do sum ficaria:
#print(f'soma: {np.sum(numero1, numero2)}')

#subtrair
print(f'subtração: {np.subtract(numero1,numero2)}')

#multiplicar
print(f'multiplicação: {np.multiply(numero1, numero2)}')

#dividir:
print(f'divisão: {np.divide(numero1, numero2)}')
```

```python
while True:
  # while True não necessita de uma condição para
  # início ou fim
  # -----
  # para finalizar o while True usamos o BREAK
  print(' ---- CALCULADORA ----')
  print('1 - somar')
  print('2- subtrair')
  print('3 - multiplicar')
  print('4 - dividir')
  print('0 - sair')

  op = input('escolha a opção: ')

  if op == "0":
    break
  elif op == "1":
    print(soma)
  elif op == "2":
    print(subtracao)
  elif op == "3":
    print(multi)
  elif op == "4":
    print(divi)
  else:
    print("escolha uma opção válida")
```

```python
#usando sistema de calculadaora - caixa de supermercado
# quais elementos tem o pagamento?

# temos produtos
# temos quantidades
# temos um total
lista_produtos=[]
lista_valor=[]
lista_quantidade=[]
#lista_total=[] - exemplo caso queira fazer valor total sem numpy

op_tem = input('Tem algum produto para passar? S/N: ')
if op_tem.lower() == 'n':
  print('Sistema finalizado')
elif op_tem.lower() == 's':
  while True:
    nome = input('Qual o produto? ')
    valor = float(input('Qual o valor? '))
    quantidade = float(input('Qual a quantidade? '))
    #total = valor*quantidade - exemplo caso queira fazer valor total sem numpy

    lista_produtos.append(nome)
    lista_valor.append(valor)
    lista_quantidade.append(quantidade)
    #lista_total.append(total) - exemplo caso queira fazer valor total sem numpy

    op_novo_produto = input('Tem outro produto? S/N: ')
    if op_novo_produto.lower() == 'n':
      print('Imprimindo o cupom fiscal...')
      print('='*20)
      print('Produto | Valor | Quantidade | Total')

      np_valor = np.array(lista_valor) #transformando a lista em numpy
      np_quantidade = np.array(lista_quantidade) # transformando a lista em np
      total = np.multiply(np_valor,np_quantidade) #calculando o valor total
#o comando zip, compacta os arrays dentro de um lugar só
      for i,j,k,l in zip(lista_produtos,lista_valor,lista_quantidade,total):
        print(f'{i} -- {j} -- {k} -- {l}')
      break
    else:
      print('Informe o novo produto')

```
