```python
#Objetivo desse programa é mostrar uma prateleira de mercado, adicionar itens nela com seus respectivos preços.

#Etapas que iremos realizar
#1 - criar um loop (while) que pergunte ao usuario se ele quer adicionar uma fruta na prateleira
#2 - a prateleira só pode ter 7 produtos
#3 - sempre que eu adicionar um produto, mostrar todos seus valores


#o comando len(variavel) mostra a quantidade de elementos dentro de algo
#vamos mostrar quantos itens já existem na prateleira
prateleira=[]
print(len(prateleira))

#vamos definir uma variável de encerramento para quando quisermos parar de adicionar frutas na prateleira
encerrar = 'não'

#condição da prateleira ter menos que 7 frutas E não quer que encerre o processo de adição
while len(prateleira) < 7 and encerrar=='não':


#pergunta se quer adicionar mais frutas. Criada a variável "adicionar" com resposta sim ou não
  adicionar=input('Deseja adicionar outra fruta? Sim ou Não: ').lower().strip()

#Se for adicionar mais frutas, perguntar qual fruta adicionar, adicioná-la na variável e depois mostrar a prateleira.
  if(adicionar=='sim'):
    fruta=input('Digite a fruta que deseja adicionar: ')
    prateleira.append(fruta) #.append() adiciona a variável dentro do array. Nesse caso adicionou 'fruta' a 'prateleira'
    print('Agora a prateleira tem as seguintes frutas:')
    for frutas in prateleira: #quando utilizo a função for, ele vai percorrer todo conteúdo dentro da lista.
      print(frutas) #nesse exemplo, o for vai trazer a primeira fruta, depois a segunda, depois a terceira e assim por diante.

#Se não for adicionar fruta no momento, perguntar se quer encerrar a adição do produto.
  else:
    print('Fruta não adicionada')
    encerrar=input('Deseja encerrar a adição de frutas? Sim ou não: ').lower().strip()

#se as condições se encerrarem, aparecer a seguinte mensagem.
print('Adição de frutas finalizada')
print(f'Sua prateleira tem {len(prateleira)} frutas')

#menu de prateleira
print('='*20)
print('A prateleira tem as seguintes frutas:')
for i in prateleira:
  print(i)
```

```python
# usando FOR

# criar um programa que pergunte ao usuário o nome de 3 músicas
# adicionar elas em uma tabela e depois
# mostrar todas elas da seguinte maneira
# a música 1 é NOME_DA_MUSICA

lista_musica=[]

print('='*30)

while len(lista_musica)<3:
  nome_musica=input(f'Digite o nome da {len(lista_musica)+1}ª música: ')
  lista_musica.append(nome_musica)

print('='*30)
print('A lista das suas músicas favoritas são:')

#aqui precisamos utilizar a função enumerate, pois ela entrega o número da posição (indice) e o conteúdo (musica)
for indice, musica in enumerate(lista_musica):

  print(f'A música {indice+1} é {musica} ')
```
