```python
# code a fizzbuzz, neste exercício iremos verificar que a ordem do if, elif, importam no resultado final
numero = int(input('Informe um número: '))

if numero < 100:
  print('Número muito pequeno')
  numero = 100

else:
  print('Sistema pronto para iniciar')

#sempre que o número é divisível por 3 mostra Fizz
#sempre que é divisível por 5 mostra Buzz
#quando é divisível por 3 e por 5 mostra FizzBuzz

for i in range(numero):

#é importante mostrar que a ordem do if, nesse caso se o número é divisível por 3 e por 5 deve vir primeiro
#caso as outras viessem primeiro, um número igual a 15, que daria verdadeiro para as 3 situações, apareceria apenas para o primeiro IF
  if(i+1)%3==0 and (i+1)%5==0:
    print(f'{i+1} é FizzBuzz')

  elif(i+1)%3==0:
    print(f'{i+1} é Fizz')

  elif(i+1)%5==0:
    print(f'{i+1} é Buzz')

  else:
    print(f'{i+1} é -')
```
