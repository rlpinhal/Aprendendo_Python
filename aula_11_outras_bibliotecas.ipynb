```python
#brute force - trabalha a tentativa e erro via combinações

import itertools #biblioteca que faz combinações
import string #preciso saber como o pc pensa em um texto
import time #computador fala para mim quanto tempo leva para descobrir uma senha

# passo 1 - alvo - o que o pc tenta adivinhar
def simulacao():
  senha_alvo = input('Digite uma senha: ')

# passo 2 - armazenas tudo  que ja foi feito
# o pc usa uma tabela ASCII para salvar os caracteres
  caracteres = string.ascii_lowercase + string.digits
  # estamos juntando todas as letras e numeros
  print('tentando descobrir sua senha!')

  tempo_inicio = time.time()
  #time.time() registra o momento que o sistema iniciou

  tentativas = 0
  encontrado = False

# passo 3 - tentar descobrir algo
  for comprimento in range(1, len(senha_alvo)+1):
    if encontrado: # if nome_encontrado ---- if var = True
      break

# passo 4 - combinado
    for combinacao in itertools.product(caracteres, repeat=comprimento):
      tentativas += 1
      palpite = "".join(combinacao)

# passo 5 - feedback - mostrar o que está fazendo
      if tentativas % 1000000 == 0:
        print(f'Tentativa de número {tentativas}')
        print(f'Testando o palpite {palpite}')


      if palpite == senha_alvo:
        tempo_fim = time.time()
        duracao = tempo_fim - tempo_inicio
        encontrado = True
        print(f'O tempo foi de {duracao}')
        print(f'Sua senha é {palpite}')
# passo 6 - finalizar tudo
simulacao()
```
