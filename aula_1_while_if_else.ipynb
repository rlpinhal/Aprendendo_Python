```python
# criando variaveis para o jogo de par ou ímpar. Iremos utilizar as funções While, if e else

#perguntar o que o jogador 1 vai escolher. Usar while para que a escolha de par ou impar esteja correta
#o código .lower() garante que tudo escrito será minúsculo e o .strip() remove espaços em branco, por exemplo "par " ficará "par"
#not in é a função que define: enquanto a variável NÃO FOR IGUAL ao que está estabelecido, continue
escolha1 =''
while escolha1 not in ['par', 'impar']:
    escolha1 = input('Jogador1 escolha par ou impar: ').lower().strip()

    if escolha1 not in ['par','impar']:
            print('Opçao inválida')

#pedir que os jogadores falem seus números. Utiliza o int, pois o valor guardado será um número inteiro
jogador1 = int(input('Jogador 1 informe um número: '))
jogador2 = int(input('Jogador 2 informe um número: '))

# uma funcao matematica interessante de saber é o módulo que é %, por exemplo 3%2 = 1 - traz o resto

# criar variável para soma dos números informados e uma variável que armazena se o resultado é par ou impar

numero = jogador1 + jogador2
resultado = ''

# lógica, vai ser utilizado condicional, if e else. Nesse caso, SE o resto da divisão da variável por 2 for diferente de 0, então o número é impar.
# Else - Senão, é par.
if numero%2 != 0:
    resultado = 'impar'
else:
    resultado = 'par'

if (escolha1 == resultado):
        print(f'Jogador 1 ganhou porque escolheu {escolha1}')
else:
        print(f'Jogador 2 ganhou porque escolheu {resultado}')
```
