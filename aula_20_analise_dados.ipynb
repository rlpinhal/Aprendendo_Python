```python
#bibliotecas de uso padrão
import pandas as pd
#biblioteca de analise
import numpy as np
#biblioteca de funções matemáticas
import matplotlib.pyplot as plt
#biblioteca para gerar gráficos
```

```python
# os dados eles estão sujos
# carregar esses dados
df_dados_oscar = pd.read_csv('/content/exercicio_qualidade_dados_oscar.csv', on_bad_lines='warn')

#tentar ver dados
display(df_dados_oscar.head(20))
```

```python
# ver onde tem dado nulo e seus tipos
df_dados_oscar.info()
```

```python
# criar contagens
print(df_dados_oscar['Vencedor'].value_counts)
# dataframe utilizado [coluna].value_counts
```

```python
#remover tudo que é duplicado
#guardo no sistema o valor atual de linhas

linhas_antes = len(df_dados_oscar)
#remover duplicatas
#para remover, você chama a base de dados fala que ela recebe ela mesma com o
#atributo drop_duplicate()

df_dados_oscar = df_dados_oscar.drop_duplicates()
```

```python
display(df_dados_oscar.head(20))
```

```python
#mostrar valores unicos
print(df_dados_oscar['Vencedor'].unique())
```

```python
#Vamos converter todos os dados para True ou False apenas
mapa_vencedores = {
    'Yes':True, 'Y': True, 'Sim': True, 'Winner!':True, '1':True,'True':True,
    'No':False, 'N':False,'0':False,'nan':False
}

#trocando os dados
# chamamos a base de dados com a coluna, fazemos ela receber ela mesmo com a
# coluna e usamos o atributo map para mapear os novos dados

df_dados_oscar['Vencedor'] = df_dados_oscar['Vencedor'].map(mapa_vencedores)
#para eliminar um dado nulo, vazio ou NaN
#chamamos a base de dados com a coluna, fazemos ela receber ela mesma com a coluna
# e usamos o atributo fillna('Valor desejado')
df_dados_oscar['Vencedor'] = df_dados_oscar['Vencedor'].fillna(False)

display(df_dados_oscar.head(20))
```

```python
#corrigir dados escritos errados
#chamamos a base de dados, com a coluna, fazemos ela receber ela mesma com a coluna e usamos
# o método replace('antes, depois')
df_dados_oscar['Indicado']=df_dados_oscar['Indicado'].replace({'Oppenheimmer':'Oppenheimer'})
display(df_dados_oscar.head(2))
```
