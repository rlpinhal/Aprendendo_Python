```python
#1 - criar uma lista com nome de 5 pets

nomes=[]
raca=[]
for i in range(5):
  nome_digitado=input(f"Digite o {i+1} nome: ")
  nomes.append(nome_digitado)
  raca_digitada=input(f"Digite a raça:")
  raca.append(raca_digitada)
```

```python
#2 - criar uma função para exibir todos os pets

print(nomes)
print(raca)
```

```python
#3 - criar uma função para cadastrar um novo pet

def cadastrar_pet():
  novo_nome=input('Digite o nome do pet: ')
  nomes.append(novo_nome)
  novo_raca=input('A raça do pet é:')
  raca.append(novo_raca)

def remover_pet():
  nome_removido=input('Digite o nome a ser removido: ')
  nomes.remove(nome_removido)
```

```python
#4 - criar um menu de opções para ver ou cadastrar pets

print('-----Lista de Pets------')
print('Escolha a opção desejada')
print('1 - Ver lista de Pets')
print('2 - Cadastrar novo Pet')
print('3 - Remover Pet')

opcao = int(input('Informe a opção desejada: '))

if opcao == 1:
  print(f'Os nomes dos pets são {nomes}')
  print(f'A raça dos pets são {raca}')

elif opcao ==2:
  cadastrar_pet()
  print('Adicionado com sucesso')

elif opcao ==3:
  remover_pet()
  print(f'Lista atualizada: {nomes}')

else:
  print('Opção Inválida')
```
